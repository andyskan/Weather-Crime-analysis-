```python
import pandas as pd
```

```python
import numpy as np
```

```python
crimedata=pd.read_csv('crimes.csv')
```

```python
crimedata.count()
```

```python
crimedata[pd.isnull(crimedata['CrmCd.Desc'])].count()#check null CrmCd.Desc
```

```python
crimedata=crimedata[pd.notnull(crimedata['CrmCd.Desc'])]#remove all null of CrmCd.Desc
```

```python
crimedata[pd.notnull(crimedata['Cross.Street'])].count()#cant be removed then with this much missing data
```

```python
crimedata[pd.isnull(crimedata['Location.1'])].count()#4397 out of 1 million, can be removed
```

```python
crimedata=crimedata[pd.notnull(crimedata['Location.1'])]
```

```python
crimedata.count()#now all data is quite good
```

```python
crimedata['diff']=pd.to_datetime(crimedata['Date.Rptd'])-pd.to_datetime(crimedata['DATE.OCC'])
```

```python
crimedata=crimedata[crimedata['diff']>='0 days']#removing OCC>Rptd
```

```python
crimedata=crimedata.reindex(np.random.permutation(crimedata.index))
```
